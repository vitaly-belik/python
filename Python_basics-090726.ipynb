## Epidemics on networks

```python
import random
from typing import List

import matplotlib.pyplot as plt
import networkx as nx
import numpy as np


# --------------------------------------------------------------------
# Parameters (exactly the same values that were used in R)
# --------------------------------------------------------------------
N = 30          # number of vertices ("graphsize")
M = 55          # number of edges to sample
T = 8           # maximal observation time (this is hardly coded below)
alpha = 0.3     # infection probability per time unit

np.random.seed(1433)         # deterministic random numbers, just like set.seed()
# random.seed(111)

# --------------------------------------------------------------------
# Build a random edge list ------------------------------------------------
# --------------------------------------------------------------------
# In R vertices are labelled 1…N.  networkx uses 0‑based indexing.
u = np.random.choice(np.arange(N), size=M, replace=True)   # heads
v = np.random.choice(np.arange(N), size=M, replace=True)   # tails

print("Number of *unique* vertices that appear in the edges:",
      len(set(u.tolist()).union(v.tolist())))


# --------------------------------------------------------------------
# Create the graph ---------------------------------------------------------
# --------------------------------------------------------------------
G = nx.Graph()
G.add_nodes_from(range(N))
# networkx Graph automatically removes self‑loops and keeps only one
# copy of each duplicate edge, so we do not need to call a “simplify” routine.
edge_list = list(zip(u.tolist(), v.tolist()))
G.add_edges_from(edge_list)

# We find the connected component CC of the network G

comp = max(nx.connected_components(G), key=len)
CC = G.subgraph(comp).copy()

# Position (layout) for all plots – identical for every time step
pos = nx.spring_layout(CC, seed = 42)


# --------------------------------------------------------------------
# Initialise infection vector: 1 for infected, 0 for susceptible ----------
# --------------------------------------------------------------------
x = np.zeros(len(G), dtype = int)
x[0] = 1          # node 1 (index 0) is infected at t=0
print("Initial infection vector x:")
print(x)

# Keep a history of the number of infected nodes.
inf_counts: List[int] = [int(x.sum())]



# --------------------------------------------------------------------
# Simulation loop ---------------------------------------------------------
# --------------------------------------------------------------------
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

axes = axes.ravel()          # flatten the array for easier indexing

times = [t for t in range(1, T + 1)]

for t in range(1, T + 1):
    # ---- Plot current state ------------------------------------------
    colors = ['red' if x[node] else 'skyblue' for node in CC.nodes]
    nx.draw_networkx(
        CC,
        pos=pos,
        ax=axes[t - 1],
        node_color = colors,
        node_size = 300,
        with_labels = True
    )
    axes[t - 1].set_title(f"t = {t}")
    axes[t - 1].axis('off')

    # ---- Update infection state ---------------------------------------
    x_next = x.copy()
    for n, m in CC.edges():
        if x[n] == 1 and x[m] == 0:
            if random.random() < alpha:
                x_next[m] = 1
                print(f"t={t}: node {m} got infected from {n}")
        elif x[m] == 1 and x[n] == 0:
            if random.random() < alpha:
                x_next[n] = 1
                print(f"t={t}: node {n} got infected from {m}")

    # Commit the new state, record statistics.
    x = x_next
    inf_counts.append(int(x.sum()))

# --------------------------------------------------------------------
# Final plot adjustments & output ---------------------------------------
# --------------------------------------------------------------------
plt.tight_layout()
plt.show()

print("\nTime series of infected counts:")
for t, cnt in enumerate(inf_counts):
    print(f"t={t:2d}  infected = {cnt}")
```
